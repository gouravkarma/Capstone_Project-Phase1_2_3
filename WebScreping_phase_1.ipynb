```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://us.puma.com/us/en/men/shop-all-mens')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
soup
```

```python
shoe_name=soup.find('h3',class_="w-full mobile:text-sm mobile:pr-0 font-bold text-base pr-5 line-clamp-2")
shoe_name
```

```python
shoe_name.text
```

```python
shoe_name=[]
color=[]
price=[]
Review=[]
```

```python
R=soup.find_all('h3',class_="w-full mobile:text-sm mobile:pr-0 font-bold text-base pr-5 line-clamp-2")
for i in R [0:10]:
    shoe_name.append(i.text)

R=soup.find_all('span',class_="group-hover:hidden group-focus-within:hidden text-puma-black-300 uppercase text-xs whitespace-nowrap")
for i in R [0:10]:
    color.append(i.text)
    
R=soup.find_all('span',class_="whitespace-nowrap text-base font-bold override:opacity-100")
for i in R [0:10]:
    price.append(i.text)

R=soup.find_all('span',class_="ml-1 pl-px text-sm")
for i in R [0:10]:
    Review.append(i.text)
```

```python
print(len(shoe_name),len(color),len(price),len(Review))
```

```python
import pandas as pd
df=pd.DataFrame({'Shoe':shoe_name,'Color':color,'Price':price,'Review':Review})
df
```

```python
df.to_csv('puma.csv')
```

```python
R=soup.find_all('a',class_="link js-tile-link")
for i in R [0:10]:
    shoe_name.append(i.text)

R=soup.find_all('div',class_="product-rating__count pl-1")
for i in R [0:10]:
    Review.append(i.text)
```
